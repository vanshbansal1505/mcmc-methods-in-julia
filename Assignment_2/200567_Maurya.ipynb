## Q2: 
The proposal function is

$f(x_1,x_2,x_3,..,x_p)=\frac{I(|x_1|\le1)I(|x_2|\le1)I(|x_3|\le1)..I(|x_p|\le1)}{2^p} $

This finally gives us that the vector x should satisfy $||x||\le1$


```julia
print("Input p: \n")
p = parse(Int, readline())
X=[]
for i in 0:999
    a=0
    while a==0
        uarr=rand(Uniform(-1, 1),p)
        if sum(uarr.^2)<1
            a=1
            push!(X,uarr)
        end
    end
end
plot(X)
```

## Q3: 
$g(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^2}{2}} $

$c=\frac{1}{\Phi(a)-\Phi(-a)} $

$\frac{f(x)}{cg(x)}=I(|x|<a)$

```julia
for i in 0:1   
    print("Input a: \n")
    ai = parse(Float64, readline())
    X3=[]
    for i in 0:10000
        acc=0
        while acc==0
            k=rand(Uniform(-ai,ai))
            k=pdf(Normal(),k)
            if abs(k)<ai
                acc=1
                push!(X3,k)
            end
        end
    end
    display(histogram(X3))
end
```
